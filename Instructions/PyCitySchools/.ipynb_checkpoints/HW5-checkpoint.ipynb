# PyCity Schools Analysis

 As a whole, schools with higher budgets, did not yield better test results. By contrast, schools with higher spending per student actually
 (\$645-675) underperformed compared to schools with smaller budgets (<\$585 per student).

* As a whole, smaller and medium sized schools dramatically out-performed large sized schools on passing math performances (89-91% passing vs 67%).

* As a whole, charter schools out-performed the public district schools across all metrics. However, more analysis will be required to glean if the effect is due to school practices or the fact that charter schools tend to serve smaller student populations per school.

```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import locale as lc
from functools import reduce

# File to Load (Remember to Change These)
school_data_to_load = "schools_complete.csv"
student_data_to_load = "students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

#rename names syntax
#district_data_complete=district_data_complete.rename(columns={'budget':'Total School Budget','size':'Total Students'})

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
#school_data_complete.head()
```

```python
#District Summary
Total_School=school_data.school_name.count()
Total_Students=student_data.student_name.count()
Total_Budget=school_data.budget.sum()
Average_Math_Score=round(student_data.math_score.mean(),2)
Average_Reading_Score=round(student_data.reading_score.mean(),2)
Overall_Passing_Rate = round((Average_Math_Score+Average_Reading_Score)/2,2)
Passing_Math_df = student_data[student_data.math_score>=70]
Passing_Read_df = student_data[student_data.reading_score>=70]
Passing_math=round(Passing_Math_df.student_name.count()/Total_Students,4)
Passing_read=round(Passing_Read_df.student_name.count()/Total_Students,4)

# print("Total School", Total_School)
# print("Total Students", Total_Students)
# print("Total Budget", '${:,.2f}'.format(Total_Budget))
# print("Average Math Score", Average_Math_Score,"%")
# print("Average Reading Score", Average_Reading_Score,"%")
# print("% Passing Math", Passing_math*100,"%")
# print("% Passing Reading", Passing_read*100,"%")
# print("% Overall Passing Rate", Overall_Passing_Rate,"%")



#Passing_math,Passing_read
```

```python
District_Summary = pd.DataFrame()
District_Summary = pd.DataFrame({'Total Schools':Total_School}, index=[0])
District_Summary['Total Students'] = '{:,.0f}'.format(Total_Students)
District_Summary['Total Budget'] = '${:,.0f}'.format(Total_Budget)
District_Summary['Average Math Score'] = Average_Math_Score
District_Summary['Average Reading Score'] = Average_Reading_Score
District_Summary['% Passing Math'] = Passing_math*100
District_Summary['% Passing Reading'] = Passing_read*100
District_Summary['% Overall Passing Rate'] = Overall_Passing_Rate
District_Summary = District_Summary.set_index('Total Schools')
#District_Summary.loc[1] = [0]*8
District_Summary.head()

#District Summary Final
```

```python
#SSG = SSG.set_index('school_name')

#Math Scores by Grade
SSG = student_data.filter(['school_name','grade','reading_score','math_score'])

#Math Scores by Grade take average by school by grade
SSG_M = round(SSG.groupby(['school_name','grade'])['math_score'].mean(),2)
SSG_R = round(SSG.groupby(['school_name','grade'])['reading_score'].mean(),2)

#Reset Series to DF and reset index for .Pivot
# when staggered the lower fields are indices ** only in notebook
SSG_MT = pd.Series.to_frame(SSG_M).reset_index()
SSG_RT = pd.Series.to_frame(SSG_R).reset_index()

#Create Pivots  rename axis remmoves grade from axis
SSG_MT_Final = SSG_MT.pivot(index='school_name', columns='grade', values ='math_score').reset_index().rename_axis(None, axis=1)
SSG_RT_Final = SSG_RT.pivot(index='school_name', columns='grade', values ='reading_score').reset_index().rename_axis(None, axis=1)


#Reorder columns
SSG_MT_Final2 = SSG_MT_Final[['school_name','9th', '10th', '11th', '12th']].copy()
SSG_RT_Final2 = SSG_RT_Final[['school_name','9th', '10th', '11th', '12th']].copy()

#Final Format

SSG_MT_Final3 = SSG_MT_Final2.set_index('school_name').rename_axis(None,axis=0)
SSG_RT_Final3 = SSG_MT_Final2.set_index('school_name').rename_axis(None,axis=0)
```
