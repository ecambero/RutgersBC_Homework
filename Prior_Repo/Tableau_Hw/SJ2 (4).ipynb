```python
data5['value'] = data5['value'].astype(str)

#Create two columns with repeated values
Forecast_Type= []
Date= []
Toss =[]

Months=['Jan', 'Feb','Mar', 'Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
numrows,_=data5.shape #since I am not interested in the second value I put a _
add=False

for row in range(numrows): #0, numrows-1
#     print(data4.loc[row])
    month_bool=[month in data5['value'][row] for month in Months]
    if True in month_bool:
       add=True
       fixed_date=data5['value'][row] #row=i
       forecast_in=row+1
       fixed_forecast=data5['value'][forecast_in]

    if add ==True:
        Date.append(fixed_date)
        Forecast_Type.append(fixed_forecast)
        
        
data5['Date']=Date
data5['Forecast_Type']=Forecast_Type
```

```python
data6=data5
```

```python
data6.head(5)
```

```python
# Clean up bad rows in data
data6['var_new']=data6['variable'].str.split('.').str[0]
data6.drop('variable',axis=1,inplace=True)
data7=data6[data6['Tableau Geo']!=('Tableau Geo Tableau Geo') ]
data8=data7[data7['Tableau Geo']!=('Tableau Geo') ]
```

```python
data8.rename(columns={'var_new':'Metric', 'value':'TM1','Forecast_Type':'Scenario'},inplace=True)
data9= data8[['Date','Tableau Geo', 'Tableau LOB', 'TM1 Geography', 'Account','Metric', 'Scenario','TM1']]
data9.head()
```

```python
# Split DF on BB and Fraud 
dataFF = data9[data9['Metric']==('Provision for Loss - Fraud')]
dataBB = data9[data9['Metric']!=('Provision for Loss - Fraud')]

#rename columns
dataFF.rename(columns={'TM1':'Provision for Loss - Fraud'}, inplace=True)
dataBB.rename(columns={'TM1':'Billed Business'}, inplace=True)

dataFF.head(), dataBB.head()
```

```python
# print(type(dataFF))
data_frames = [dataFF,dataBB ]

#merge as needed
Final_Data2 = reduce(lambda  left,right: pd.merge(left,right,on=['Date', 'Tableau Geo', 'Tableau LOB','TM1 Geography','Account', 'Scenario' ],
                                            how='outer'), data_frames).fillna(0)
Final_Data2.head()
```

```python
#Drop Metric column for Final prep 
Final_Data2.drop(['Metric_x','Metric_y'],axis=1,inplace=True)
# Final_Data2.drop('TM1 Geography',axis=1,inplace=True)
```

```python
Final_Data2.set_index('Date', inplace=True)
Final_Data2.head(5)
```
