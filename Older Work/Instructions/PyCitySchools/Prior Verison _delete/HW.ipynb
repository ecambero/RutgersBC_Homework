```python

# Dependencies and Setup
import pandas as pd
import numpy as np
import locale as lc
from functools import reduce

# File to Load (Remember to Change These)
school_data_to_load = "schools_complete.csv"
student_data_to_load = "students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
#District Summary
Total_School=school_data.school_name.count()
Total_Students=student_data.student_name.count()
Total_Budget=school_data.budget.sum()
Average_Math_Score=round(student_data.math_score.mean(),2)
Average_Reading_Score=round(student_data.reading_score.mean(),2)
Overall_Passing_Rate = round((Average_Math_Score+Average_Reading_Score)/2,2)
Passing_Math_df = student_data[student_data.math_score>=70]
Passing_Read_df = student_data[student_data.reading_score>=70]
Passing_math=round(Passing_Math_df.student_name.count()/Total_Students,4)
Passing_read=round(Passing_Read_df.student_name.count()/Total_Students,4)

# print("Total School", Total_School)
# print("Total Students", Total_Students)
# print("Total Budget", '${:,.2f}'.format(Total_Budget))
# print("Average Math Score", Average_Math_Score,"%")
# print("Average Reading Score", Average_Reading_Score,"%")
# print("% Passing Math", Passing_math*100,"%")
# print("% Passing Reading", Passing_read*100,"%")
# print("% Overall Passing Rate", Overall_Passing_Rate,"%")



#Passing_math,Passing_read
```

```python
District_Summary = pd.DataFrame()
District_Summary = pd.DataFrame({'Total Schools':Total_School}, index=[0])
District_Summary['Total Students'] = '{:,.0f}'.format(Total_Students)
District_Summary['Total Budget'] = '${:,.0f}'.format(Total_Budget)
District_Summary['Average Math Score'] = Average_Math_Score
District_Summary['Average Reading Score'] = Average_Reading_Score
District_Summary['% Passing Math'] = Passing_math*100
District_Summary['% Passing Reading'] = Passing_read*100
District_Summary['% Overall Passing Rate'] = Overall_Passing_Rate
District_Summary = District_Summary.set_index('Total Schools')
#District_Summary.loc[1] = [0]*8
District_Summary.head(10)

#District Summary Final
```

```python
#District Summary
# Create an overview table that summarizes key metrics about each school, including:
  # School Name
  # School Type
  # Total Students
  # Total School Budget
  # Per Student Budget
  # Average Math Score
  # Average Reading Score
  # % Passing Math
  # % Passing Reading
  # Overall Passing Rate (Average of the above two)
  
# Create a dataframe to hold the above results

SS = student_data.filter(['school_name','reading_score','math_score'])

#Mean Score for all schools
SS_Mean = SS.groupby('school_name')['reading_score','math_score'].mean()

#All schools w\ scores >=70 by school_name
SS_M = SS[SS['math_score']>=70][['school_name','math_score']]
SS_R = SS[SS['reading_score']>=70][['school_name','reading_score']]

# as index=False converts a series into a DF which can be merged
SS_R1= SS_R.groupby('school_name', as_index=False)['reading_score'].count()
SS_M1= SS_M.groupby('school_name', as_index=False)['math_score'].count()

#SS_M= SS_V2.groupby('school_name')[(SS['math_score']>=70)].count()



# compile the list of dataframes you want to merge
data_frames = [school_data, SS_R1, SS_M1, SS_Mean]

#merge as needed
district_data_complete = reduce(lambda  left,right: pd.merge(left,right,on=['school_name'],
                                            how='left'), data_frames)

#district_data_complete.head(5)                                                              
                                                             
#district_data_complete2 = district_data_complete.rename(columns={'reading_Score':'Passing_Reading_Score'}, inplace=True)


#district_data_complete.rename(index=str,columns={'School ID':'S'}) 

#School_Summary = pd.DataFrame({'Total Schools':Total_School}, index=[0])

```
