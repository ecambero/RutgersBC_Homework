```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "city_data.csv"
ride_data_to_load = "ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
data_complete = pd.merge(city_data, ride_data, how="left", on=["city", "city"])
# Display the data table for preview
```

```python
data_complete.head(10)
```

```python
avg_fare= round(data_complete.groupby(['city','type'], as_index=False)['fare'].mean(),2)
total_rides= round(data_complete.groupby(['city','type'], as_index=False)['ride_id'].count(),2)
ride_per = data_complete.groupby(['city','type'], as_index=False)['driver_count'].min()
```

```python
data_complete.type.unique()
```

```python
# # compile the list of dataframes you want to merge
# data_frames = [total_rides, avg_fare,ride_per]

# #merge as needed
# scatter_data= reduce(lambda  left,right: pd.merge(left,right,on=['city','type'],
#                                             how='left', data_frames))


# compile the list of dataframes you want to merge
data_frames = [total_rides, avg_fare, ride_per]

#merge as needed
scatter_data = reduce(lambda  left,right: pd.merge(left,right,on=['city','type'],
                                            how='left'), data_frames)

#Add Colors
scatter_data.loc[scatter_data.type=='Urban', 'Color'] = 'coral'
scatter_data.loc[scatter_data.type=='Suburban', 'Color'] = 'skyblue'
scatter_data.loc[scatter_data.type=='Rural', 'Color'] = 'gold'

#scatter_data.head()

#Subset data for color formatting
rur_df =scatter_data.loc[scatter_data.type=='Rural']
urb_df =scatter_data.loc[scatter_data.type=='Urban']
sub_df =scatter_data.loc[scatter_data.type=='Suburban']

```

## Bubble Plot of Ride Sharing Data

```python
# Build the scatter plots for each city types

plt.xlim(1,40)
plt.ylim(20,40)

#Urban

urb2_df = urb_df.groupby(['city'])

u_avg_fare= urb2_df['fare'].mean()
u_total_rides=urb2_df['ride_id'].sum()
u_ride_per=urb2_df['driver_count'].min()

u_x = u_total_rides
u_y = u_avg_fare
u_s = u_ride_per
u_t = urb2_df['Color']

plt.scatter(u_x, u_y, s=10*u_s, color = 'coral', alpha=0.3, label='urban')


#Suburban

sub2_df = sub_df.groupby(['city'])

s_avg_fare= sub2_df['fare'].mean()
s_total_rides=sub2_df['ride_id'].sum()
s_ride_per=sub2_df['driver_count'].min()

s_x = s_total_rides
s_y = s_avg_fare
s_s = s_ride_per
s_t = sub2_df['Color']

plt.scatter(s_x, s_y, s=10*s_s, color = 'skyblue', alpha=0.3,marker='o', label='suburban')

# ###define dictionary 

# # Obtain the x and y coordinates for each of the three city types

# #Rural

rur2_df = rur_df.groupby(['city'])

r_avg_fare= rur2_df['fare'].mean()
r_total_rides=rur2_df['ride_id'].sum()
r_ride_per=rur2_df['driver_count'].min()

r_x = r_total_rides
r_y = r_avg_fare
r_s = r_ride_per
r_t = rur2_df['Color']


plt.scatter(r_x, r_y, s=10*r_s, color = 'gold', alpha=0.3,marker='o', label='rural')

# Incorporate the other graph properties

plt.tight_layout()
plt.grid()
plt.title('Pyber Ride Sharing Data for 2016')
plt.xlabel('Total Number of Rides (Per City)')
plt.ylabel('Average Fare')

# Create a legend

plt.legend(loc='best',title='City Types')

# Incorporate a text label regarding circle size

plt.text(41,35,"Note:\nCircle size correlates with driver count per city. ")

# Save Figure

plt.savefig("Pyber_Ride_Sharing_Data_for_2016.png")


```
